```python
import pandas as pd
import os
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
coop_dirs = ['PairedTestingSessions', 'Training_COOP_Fiber_new', 'Training_COOPERATION']
root = '/gpfs/radev/pi/saxena/aj764'
```

```python
df = pd.DataFrame(columns = ['dir', 'session', 'video', 'color type', 'color pair', 'tracked', 'error', 'lever', 'mag', 'event id', 'fiber pho', 'coop thresh', 'animal id', 'cohort', 'visual', 'vehicle', 'familiar'])

for coop in coop_dirs:
    sessions = os.listdir(f'{root}/{coop}')
    for sesh in sessions:
        if os.path.exists(f'{root}/{coop}/{sesh}/Videos/'):
            vids = os.listdir(f'{root}/{coop}/{sesh}/Videos/')
            vids = [x for x in vids if '.mp4' in x]
            for vid in vids:
                if coop != 'PairedTestingSessions' or (coop == 'PairedTestingSessions' and 'Coop' in vid):
                    try:
                        name = vid.split('.')[0]
                        date = name.split('_')[0]
                        color_type = 'dyed' if (int(date[-2:]) == 24 and int(date[:2]) < 11) else 'collar'
                        color = name.split('_')
                        animal_id = [x for x in color if (len(x) == 13 or len(x) == 14)][0]
                        color = animal_id.split('-')
                        color = [color[0][-1], color[1][-1]]
                        color_pair = ''
                        if 'R' in color:
                            color_pair += 'R'
                        if 'G' in color:
                            color_pair += 'G'
                        if 'Y' in color:
                            color_pair += 'Y'
                        if 'B' in color:
                            color_pair += 'B'
                        preds = os.path.exists(f'{root}/{coop}/{sesh}/Tracking/slp/{name}.predictions.slp')
                        if preds:
                            with h5py.File(f'{root}/{coop}/{sesh}/Tracking/h5_uncorrected/{name}.predictions.h5','r') as f:
                                locations = f["tracks"][:].T
                            error = np.sum(np.isnan(locations)) / np.prod(locations.shape)
                        else:
                            error = np.nan
                        lever = os.path.exists (f'{root}/{coop}/{sesh}/Behavioral/processed/lever/{name}_lever.csv')
                        mag = os.path.exists(f'{root}/{coop}/{sesh}/Behavioral/processed/mag/{name}_mag.csv')
                        if lever:
                            levs = pd.read_csv(f'{root}/{coop}/{sesh}/Behavioral/processed/lever/{name}_lever.csv')
                            event_id = 'RatID' in levs.columns.tolist()
                            coop_thresh = np.average(levs['CoopTimeLimit'])
                        else:
                            event_id = False
                            coop_thresh = np.nan
                        if coop == 'Training_COOP_Fiber_new':
                            neuronal = os.path.exists(f'{root}/{coop}/{sesh}/Neuronal/{name[:-8]}.mat')
                        else:
                            neuronal = os.path.exists(f'{root}/{coop}/{sesh}/Neuronal/{name}.mat')
                        cohort = animal_id.split('-')[0][:2]
                        visual = 'opaque' if 'opaque' in sesh else ('translucent' if 'translucent' in sesh else 'transparent')
                        pos_vehs = ['CNO', 'VEH', 'DCZ', 'SAL', 'MDMA']
                        vehicle = [x for x in pos_vehs if x in sesh]
                        if len(vehicle) == 1:
                            vehicle = vehicle[0]
                        else:
                            vehicle = 'None'
                        familiar = 'unfamiliar' if (animal_id.split('-')[0][:-1] != animal_id.split('-')[1][:-1] and cohort == 'KL') else 'familiar'
                        if cohort == 'KM' or cohort == 'KF':
                            familiar = 'unsure'
                        df.loc[len(df)] = [coop, sesh, name, color_type, color_pair, preds, error, lever, mag, event_id, neuronal, coop_thresh, animal_id, cohort, visual, vehicle, familiar]
                    except Exception as e:
                        print(e, coop, sesh, vid)

```

```python
uses = []
for index, row in df.iterrows():
    if (row['tracked'] and row['lever'] and row['mag'] and (row['error'] < 0.2)):
        use = 'control'
        if (row['vehicle'] == 'None' or row['vehicle'] == 'CNO') and row['visual'] == 'transparent' and row['familiar'] == 'familiar':
            use = 'normal'
        if row['familiar'] == 'unsure':
            use = 'unsure (familiarity)'
    else:
        use = 'unuseable'
    uses.append(use)
df['useable'] = uses
```

```python
df.to_csv('all_coop_uncorrected.csv', index=False)
# df = pd.read_csv('all_coop.csv')
```

```python
model_df = pd.DataFrame(columns = ['color type', 'color pair', 'unuseable', 'unuseable fiber']) # , 'unuseable 1', 'unuseable 1.25', 'unuseable 1.5', 'unuseable 2', 'unuseable 3', 'unuseable 4', 'unuseable 5', 'unuseable 6'])
for color_type in np.unique(df['color type']):
    for color_pair in np.unique(df['color pair']):
        unuseable = df[(df['color type'] == color_type) & (df['color pair'] == color_pair) & (df['useable'] == 'unuseable') & ((~df['tracked']) | (df['error'] > 0.2))]
        fiber = unuseable[unuseable['fiber pho']]
        
        model_df.loc[len(model_df)] = [color_type, color_pair, unuseable.shape[0], fiber.shape[0]]

```

```python
model_df
model_df.sort_values(by='unuseable', ascending=False)
model_df.sort_values(by='unuseable fiber', ascending=False)
# model_df.to_csv('model_priority_ranking.csv', index=False)
```
