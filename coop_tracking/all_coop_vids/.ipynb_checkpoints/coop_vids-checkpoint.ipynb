```python
import pandas as pd
import numpy as np
```

```python
# loads csv with all of the videos
df = pd.read_csv('all_coop_vid.csv')
```

```python
np.unique(df['color pair'])
```

```python
# CHANGE THINGS HERE!!
color_type = 'collar' # can be either 'collar' or 'dyed'
color_pair = 'RG' # can be 'GB', 'GY', 'RB', 'RG', 'RY', or 'YB'
num_vids = 10 # number of video suggestions you want to generate
vid_sort = 'order' # can be either 'rand' or 'ordered' which means ordered by the tracking error

tracked = False # means that the videos hasn't been tracked at all
error_threshold = 0.2 # any videos that has been tracked but has greater than 20% error needs to be improved
```

```python
# this shows all videos that fit the abvoe color type and color pair AND still need to be tracked or need their tracking improved
df[(df['color type'] == color_type) & (df['color pair'] == color_pair) & ((df['tracked'] == tracked) | (df['error'] > 0.2))]
```

```python
subset_df = df[(df['color type'] == color_type) & (df['color pair'] == color_pair) & ((df['tracked'] == tracked) | (df['error'] > 0.2))]
subset_df = subset_df.sort_values(by='error', ascending=False)

if vid_sort == 'rand':
    videos = subset_df.sample(n=10)
else:
    videos = subset_df.head(10)
video_names = videos['dir'] + '/' + videos['session'] + '/' + videos['video']

# prints out the file path of the videos to label
for vid in video_names:
    print(vid)
```
