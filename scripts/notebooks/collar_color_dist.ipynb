```python
# choose 4 random videos from each of the remaining colors to start labeling on!!
remain = ['RB', 'GB', 'RY', 'GY']
remain_trans = {'RB':'Red-Blue', 'GB':'Green-Blue', 'RY':'Red-Yellow', 'GY':'Green-Yellow'}
for r in remain:
    color_df = collar_df[collar_df['color pair'] == r]
    test = color_df[color_df['test/train'] == 'test']
    train = color_df[color_df['test/train'] == 'train']
    # print(r, test.shape[0], train.shape[0])
    random_test = test.sample(n=3)
    for i,row in random_test.iterrows():
        color_pair = remain_trans[row['color pair']]
        print(f"cp PairedTestingSessions/{row['session']}/Videos/{row['vid']}.mp4 Nina_Model_Testing/Collars/{color_pair}/videos/")
    
    if train.shape[0] != 0:
        random_train = train.sample(n=1)
        for i,row in random_train.iterrows():
            color_pair = remain_trans[row['color pair']]
            print(f"cp Training_COOPERATION/{row['session']}/{row['vid']}.mp4 Nina_Model_Testing/Collars/{color_pair}/videos/")
    

    # random_test = test.sample(n=2)
    # random_train = train.sample(n=2)
```
