```python
import pandas as pd
```

```python
errors = pd.read_csv('errors.csv')
```

```python
vids = []
for index, row in errors.iterrows():
    if index % 16 == 0:
        vids.append(str(row).split()[1])
```

```python
from collections import Counter
```

```python
len(Counter(vids).keys())
```

```python
Counter(vids).values()
```

```python
# key needs to be (date, id) tuple, value needs to be vid name
```

```python
vid_dict = {}
for vid in vids:
    parsed = vid.split('_')
    date = parsed[0]
    an_id = [x for x in parsed if ('KL' in x) or ('EB' in x)][0]
    if (date, an_id) not in vid_dict.keys():
        vid_dict[(date, an_id)] = []
    vid_dict[(date, an_id)].append(vid)
```

```python
for key, value in vid_dict.items():
    if len(value) == 4: 
        print(key, value)
# these are the ones that we should check that mag and lever ones aren't switched!!
```
