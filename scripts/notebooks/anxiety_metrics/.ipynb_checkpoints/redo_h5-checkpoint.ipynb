Will print necessary linux command to recreate all of the h5 files in the case that we decide didn't like the error correction (reconverts the original, untouched slp files to h5 files that can be loaded in python). 

```python
import sys
sys.path.append('../../')
```

```python
import pandas as pd
import os
```

```python
df = pd.read_csv('large_arena_preds_df.csv')
```

```python
tot = 0
done = 0
for index, row in df.iterrows():
    if index != 0:
        t = 'Training_LARGEARENA_Individual'
        s = row['session']
        v = row['vid']
        start = 'module load miniconda; conda activate sleap_old;'
        if not os.path.exists(f'/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5_not_corrected/'):
            print(f'/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5_not_corrected/')
            os.mkdir(f'/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5_not_corrected/')
        # middle = f"mkdir /gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5_not_corrected/;"
        end = f"sleap-convert --format analysis -o \"/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5_not_corrected/{v}predictions.h5\" \"/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/slp/{v}predictions.slp\""
        # print(start + middle + end)
        # print(start + end)
        if os.path.exists(f'/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5_not_corrected/{v}predictions.h5'):
            done += 1
        tot += 1
print(done / tot)

```

```python
from src.classes.vid_loader import VidLoader
from src.utils.error_utils import get_vs
```

```python
vids = VidLoader(out=False)
```

```python
def print_convert(vid_dict, trial_type):
    for key, value in vid_dict.items():
        for vid in value:
            t = 'PairedTestingSessions' if trial_type == 'test' else 'Training_COOPERATION'
            v, s = get_vs(key, vid)
            print(f"sleap-convert --format analysis -o \"/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/h5/{v}.predictions.h5\" \"/gpfs/radev/pi/saxena/aj764/{t}/{s}/Tracking/slp/{v}.predictions.slp\"")
```

```python
# if doing in the command line, need to do below command first:
# module load miniconda
# conda load sleap_old
```

```python
print_convert(vids.pts_single_vids, 'test')
print_convert(vids.pts_multi_vids, 'test')
print_convert(vids.tc_multi_vids, 'train')
```
