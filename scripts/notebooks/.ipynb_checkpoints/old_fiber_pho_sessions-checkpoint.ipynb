```python
fig, ax = plt.subplots(ncols = 2, sharex=True, sharey=True)
ax[0].set_aspect('equal')
ax[1].set_aspect('equal')
for index, row in (preds.df.iterrows()):
    vid = row['vid']
    original_error = old_df[old_df['vid'] == vid]['initial nan'].item()
    redo_error = row['initial nan']
    no_tail_error = row['initial nan no tail']
    if original_error != -1 and redo_error != -1:
        ax[0].plot(original_error, redo_error, '.')
        ax[1].plot(original_error, no_tail_error, '.')
line = np.arange(0, 1, 0.01)
ax[0].plot(line, line, 'k', linestyle='dashed')
ax[1].plot(line, line, 'k', linestyle='dashed')

# Add shared x-label and individual y-labels
ax[0].set_xlabel("color pair model initial error")
ax[0].set_ylabel("fiber model initial error")
ax[1].set_xlabel("color pair model initial error")
ax[1].set_ylabel("fiber model w/o tail error")
fig.suptitle("change in pred error for already predicted videos", y=0.80)

plt.tight_layout(rect=[0, 0, 1, 0.94])  # Leave room for adjusted title
plt.show()
```

## new (collar) fiber sessions

```python
from src.utils.global_utils import CENTROID, TOPDOWN, ROOTDIR
import h5py
from src.utils.error_utils import get_nan_prec, high_vel_nan
import tqdm
```

```python
df = pd.DataFrame(columns=['vid', 'session', 'single/multi', 'test/train', 'pred', 'color pair',
       'initial nan', 'vel nan', 'correct', 'final nan', 'levers', 'mags', 'trial type', 'familiarity', 'fiber pho'])
```

```python
sessions = os.listdir('/gpfs/radev/pi/saxena/aj764/Training_COOP_Fiber/')
```

```python
index = 0
for sesh in sessions:
    try:
        cp = sesh.split('_')
        cp = [x for x in cp if '-' in x][0]
        cp = cp.split('-')
        cp = cp[0][-1] + cp[1][-1]
        trial_key = ''
        if 'R' in cp:
            trial_key += 'R'
        if 'G' in cp:
            trial_key += 'G'
        if 'Y' in cp:
            trial_key += 'Y'
        if 'B' in cp:
            trial_key += 'B'

        files = os.listdir('/gpfs/radev/pi/saxena/aj764/Training_COOP_Fiber/' + sesh)
        vid = [x for x in files if '.mp4' in x][0]
        df.loc[index] = [vid, sesh, 'multi', 'train', False, trial_key, -1, -1, False, -1, False, False, 'coop', 'n/a', True]
        index += 1
    except:
        print(sesh)
```

```python
df.to_csv('collar_fiber_df.csv', index=False)
```

```python
# sleap tracking commands for regular collar models
start_command = "module load miniconda; conda activate sleap; cd /gpfs/radev/pi/saxena/aj764/; "
for index, row in (parsed_df.iterrows()):
    vid = row['vid']
    video_path = f"/gpfs/radev/pi/saxena/aj764/Training_COOP_Fiber/{row['session']}/{row['vid']}"
    output_path = f"/gpfs/radev/pi/saxena/aj764/Training_COOP_Fiber/{row['session']}/"

    centroid_model = CENTROID[('collar', row['color pair'])]
    topdown_model = TOPDOWN[('collar', row['color pair'])]
    output_file = vid[:-3]
    # check if slp file is there
    if not os.path.exists(output_path + output_file + 'slp' ):
        track_command = f'sleap-track "{video_path}" --first-gpu -o "{output_path + output_file + 'slp'}" -m "{centroid_model}" -m "{topdown_model}"'
        convert_command = f'; sleap-convert --format analysis -o "{output_path + output_file + 'h5'}" "{output_path + output_file + 'slp'}"'
        print(start_command + track_command + convert_command)
```

```python
# sleap tracking commands
start_command = "module load miniconda; conda activate sleap; cd /gpfs/radev/pi/saxena/aj764/; "
for index, row in (parsed_df.iterrows()): # for vid in vids:
    vid = row['vid']
    video_path = f"/gpfs/radev/pi/saxena/aj764/fiber_videos/tcf/Videos/{vid}"
    output_path = f"/gpfs/radev/pi/saxena/aj764/fiber_videos/tcf/Tracking"

    # NEWEST FIBER ONLY MODELS
    # centroid_model = '/gpfs/radev/pi/saxena/aj764/Nina_Model_Testing/Fiber/models/250813_170357.centroid'
    # topdown_model = '/gpfs/radev/pi/saxena/aj764/Nina_Model_Testing/Fiber/models/250813_170357.multi_class_topdown'
    centroid_model = '/gpfs/radev/pi/saxena/aj764/Nina_Model_Testing/Fiber/models/251012_233935.centroid'
    topdown_model = '/gpfs/radev/pi/saxena/aj764/Nina_Model_Testing/Fiber/models/251012_233935.multi_class_topdown'
    
    output_file = vid[:-3]
    track_command = f'sleap-track "{video_path}" --first-gpu -o "{output_path + '/slp/' + output_file + 'slp'}" -m "{centroid_model}" -m "{topdown_model}"'
    convert_command = f'; sleap-convert --format analysis -o "{output_path + '/h5/' + output_file + 'h5'}" "{output_path + '/slp/' + output_file + 'slp'}"'
    print(start_command + track_command + convert_command)
```

```python
# collar_model_df = pd.read_csv("collar_fiber_df_collar_model.csv")
fiber_model_df = pd.read_csv("collar_fiber_df_fiber_model.csv")
# get errors for regular tracking
for index, row in tqdm.tqdm(parsed_df.iterrows()):
    # load collar pred file
    try:
        with h5py.File(ROOTDIR + '/Training_COOP_Fiber/' + row['session'] + '/' + row['vid'][:-4] + '.h5','r') as f:
            locations = f["tracks"][:].T
        # get collar pred nan
        collar_model_df.at[index, 'initial nan'] = get_nan_prec(locations)
        locations = high_vel_nan(locations)
        collar_model_df.at[index, 'vel nan'] = get_nan_prec(locations)
        collar_model_df.at[index, 'pred'] = True
    except:
        print(row['vid'])
    
    
# get errors for fiber tracking
for index, row in tqdm.tqdm(parsed_df.iterrows()):
    # load collar pred file
    try:
        with h5py.File(ROOTDIR + 'fiber_videos/tcf/Tracking/h5/' + row['vid'][:-4] + '.h5','r') as f:
            locations = f["tracks"][:].T
        # get collar pred nan
        fiber_model_df.at[index, 'initial nan'] = get_nan_prec(locations)
        locations = high_vel_nan(locations)
        fiber_model_df.at[index, 'vel nan'] = get_nan_prec(locations)
        fiber_model_df.at[index, 'pred'] = True
    except:
        print(row['vid'])
```

```python
# get errors for regular tracking
for index, row in tqdm.tqdm(parsed_df.iterrows()):
    # load collar pred file
    try:
        with h5py.File(ROOTDIR + '/Training_COOP_Fiber/' + row['session'] + '/' + row['vid'][:-4] + '.h5','r') as f:
            locations = f["tracks"][:].T
        locations = locations[:, :-1, :, :]
        # get collar pred nan
        collar_model_df.at[index, 'initial nan no tail'] = get_nan_prec(locations)
        locations = high_vel_nan(locations)
        collar_model_df.at[index, 'vel nan no tail'] = get_nan_prec(locations)
    except:
        print(row['vid'])
    
    
# get errors for fiber tracking
for index, row in tqdm.tqdm(parsed_df.iterrows()):
    # load collar pred file
    try:
        with h5py.File(ROOTDIR + 'fiber_videos/tcf/Tracking/h5/' + row['vid'][:-4] + '.h5','r') as f:
            locations = f["tracks"][:].T
        locations = locations[:, :-1, :, :]
        # get collar pred nan
        fiber_model_df.at[index, 'initial nan no tail'] = get_nan_prec(locations)
        locations = high_vel_nan(locations)
        fiber_model_df.at[index, 'vel nan no tail'] = get_nan_prec(locations)
    except:
        print(row['vid'])
```
