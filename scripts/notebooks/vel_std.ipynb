```python
import pandas as pd
import numpy as np
import tqdm
import sys
import matplotlib.pyplot as plt
sys.path.append('../../')
```

```python
from src.utils.error_utils import high_vel_nan, lowess_fill, graph_locations, get_nan_prec, nan_good, load_file
```

```python
df = pd.read_csv('dyed_preds_df.csv')
df = df[(df['single/multi'] == 'multi') & (df['pred'] == True) & (df['initial nan'] < 0.05)]
```

```python
df.shape
```

```python
def get_max_vel(locations, node):
    x = np.sqrt(np.sum(np.square(locations[:-1, node, :, :] - locations[1:, node, :, :]), axis=1))
    mean, std = np.nanmean(x, axis=0), np.nanstd(x, axis=0)
    high, low = mean + (3 * std), mean - (3 * std)   
    return mean, std, high
    
```

```python
means = np.zeros((df.shape[0] * 2, 5))
stds = np.zeros((df.shape[0] * 2, 5))
highs = np.zeros((df.shape[0] * 2, 5))
```

```python
count = 0
for index, row in tqdm.tqdm(df.iterrows()):
    locations = load_file(row)
    for node in range(locations.shape[1]):
        mean, std, high = get_max_vel(locations, node)
        means[count:count+2, node] = mean
        stds[count:count+2, node] = std
        highs[count:count+2, node] = high
    count += 2
```

```python
print(np.mean(means, axis=0))
```
